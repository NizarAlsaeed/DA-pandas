## Part1: bike data

#### Cleaning the data

```python
#cleaning the data first; it seems that two tables were joined together and the header of the 
#second table appears as a data row
df = pd.read_csv('./archive/trip.csv',low_memory=False)
df.loc[df['trip_id'] == 'trip_id']
```

```python
#dropping that row
df = df.drop(50793,axis=0)
df.loc[df['trip_id'] == 'trip_id']
```

```python
df
```

```python
#to do calculations, first,I'll change the datatype from objects to numeric
df.dtypes
```

```python
# data-type specific converters pd.to_datetime, pd.to_timedelta and pd.to_numeric.
df['trip_id'] = pd.to_numeric(df['trip_id'],downcast='integer')
df['birthyear'] = pd.to_numeric(df['birthyear'],downcast='integer')
df['tripduration'] = pd.to_numeric(df['tripduration'],downcast='float')
df['tripduration'] = pd.to_numeric(df['tripduration'],downcast='float')

df.dtypes
```

#### Now, the data is ready

```python
#What is the average trip duration for a borrowed bicycle?
df['tripduration'].mean()
```

```python
#Whatâ€™s the most common age of a bicycle-sharer?
pd.datetime.now().year - df['birthyear'].value_counts().index[0]
```
